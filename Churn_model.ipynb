# Churn Prediction Pipeline

## Overview:
Churn is an important metric to minimize for a retailer, especially for maximizing profit. Tracking customers who are at risk of churning out of the company is essential to retain a customer base, and by utilizing machine learning algorithms a business is able to make predictions and target their customers who are showing signs of potentially churning. 

The goal of this project is to develop a model that can be implemented to detect potential churn risks that a retailer has in order to target those customers through campaigns and other marketing to draw them back into a position where they are unlikely to churn.


## Business Problem

When customers at risk of churn are able to be predicted before they churn away a company is able to save money on advertising campaigns. Retaining a customer is much more cost effective compared to constantly running wide spread advertisement campaigns to attract new customers. If a company is able to use indicators in their data from past customers who have churned vs customers who have not, it is possible to target the current customers who are behaving similar to previous customers that did churn and then target efforts on the at risk customers through campaigns and giving incentives to stay. 

The potential stakeholders are:
 
    1. CMO (Chief Marketing Officer): Use insights for retention campaigns

    2. CRM (Customer Relationship Management) Team: Team in charge of the campaigns

    3. Data Science Team: Builds, uses, and maintains predictive model

    4. Executive Leadership: Use forecasts for strategic planning

## Key Questions

1) Is there a relationship between high cash back and customer churn?

2) Does 'distance from warehouse', high shipping times vs low shipping times, affect the likelyhood of churn?

3) Does 'Satisfaction Score' being low indicate a customer will churn?

## Data Cleaning

```python
#Remove rows missing essential data, Customer number and Churn status
data = data.dropna(subset=['CustomerID', 'Churn'])  # essential columns

#Split columns into categorical and numeric for imputation
categorical_features = [
    'PreferredLoginDevice', 'CityTier', 'PreferredPaymentMode', 
    'Gender', 'MaritalStatus', 'PreferedOrderCat'
]
numeric_features = [
    'Tenure', 'WarehouseToHome', 'HourSpendOnApp', 'NumberOfDeviceRegistered',
    'SatisfactionScore', 'NumberOfAddress', 'OrderAmountHikeFromlastYear',
    'CouponUsed', 'OrderCount', 'DaySinceLastOrder', 'CashbackAmount'
]
#Due to low missing data, filling numeric columns with the median value since columns with missing data are low correlation to churn
for col in numeric_features:
    if col in data.columns:
        median_val = data[col].median()
        data[col].fillna(median_val, inplace=True)
        
print("Missing values after numeric imputation:\n", data[numeric_features].isnull().sum())
```

```python
#Audit summary
print("\nData Cleaning Audit Summary:")
print(f"Total Rows after Cleaning: {data.shape[0]}")
print(f"Total Unique Customers after Cleaning: {data['CustomerID'].nunique()}")
# If 'Country' column exists, print its nunique, else skip
if 'Country' in data.columns:
	print(f"Total Countries after Cleaning: {data['Country'].nunique()}")
print("Data check:", display(data.head(5)))
```

## Feature Engineering

```python
# Average order value
data['AverageOrderValue'] = data['OrderAmountHikeFromlastYear'] / (data['OrderCount'].replace(0,1))

# Cashback per order
data['CashbackPerOrder'] = data['CashbackAmount'] / (data['OrderCount'].replace(0,1))

# Coupon usage per order
data['CouponPerOrder'] = data['CouponUsed'] / (data['OrderCount'].replace(0,1))

# Time spent per device
data['HourPerDevice'] = data['HourSpendOnApp'] / (data['NumberOfDeviceRegistered'].replace(0,1))

# Recency flags
data['RecentActivityFlag'] = (data['DaySinceLastOrder'] <= 30).astype(int)
data['InactiveFlag'] = (data['DaySinceLastOrder'] > 60).astype(int)

# Complaint rate
data['ComplaintRate'] = data['Complain'] / (data['OrderCount'].replace(0,1))

# Satisfaction normalized by tenure
data['SatisfactionPerTenure'] = data['SatisfactionScore'] / (data['Tenure'].replace(0,1))

# Device & Payment mode
data['Device_Payment_Interaction'] = data['PreferredLoginDevice'].astype(str) + '_' + data['PreferredPaymentMode'].astype(str)

# City tier & Average order value
data['City_AOV_Interaction'] = data['CityTier'].astype(str) + '_' + pd.cut(data['AverageOrderValue'], bins=3, labels=['Low', 'Medium', 'High']).astype(str)

# Coupon usage & Satisfaction
data['Coupon_Satisfaction_Ratio'] = data['CouponPerOrder'] / (data['SatisfactionPerTenure'].replace(0,np.nan))

# Fill any NaNs from division by zero with median
numeric_features_to_fill = ['Coupon_Satisfaction_Ratio', 'CashbackPerOrder', 'HourPerDevice', 'ComplaintRate', 'SatisfactionPerTenure', 'AverageOrderValue', 'CouponPerOrder']
for col in numeric_features_to_fill:
    data[col].fillna(data[col].median(), inplace=True)

categorical_features += ['Device_Payment_Interaction', 'City_AOV_Interaction']
numeric_features += ['AverageOrderValue', 'CashbackPerOrder', 'CouponPerOrder', 'HourPerDevice',
                    'RecentActivityFlag', 'InactiveFlag', 'ComplaintRate', 'SatisfactionPerTenure',
                    'Coupon_Satisfaction_Ratio']
```

```python
print("Updated Data Shape:", data.shape)
print("Sample of engineered features:")
display(data[categorical_features + numeric_features].head())
```

## Data Split

```python
X = data[categorical_features + numeric_features]
y = data['Churn']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.25, random_state=42
)
```

## Pipeline Setup

```python
numeric_transformer = Pipeline(steps=[('scaler', StandardScaler())])
categorical_transformer = Pipeline(steps=[('onehot', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_features),
    ('cat', categorical_transformer, categorical_features)
])
```

## Modeling Strategy

```python
models = {
    "LogisticRegression": LogisticRegression(max_iter=500),
    "RandomForest": RandomForestClassifier(),
    "GradientBoosting": GradientBoostingClassifier()
}

param_grids = {
    "LogisticRegression": {'model__C': [0.1, 1, 10]},
    "RandomForest": {'model__n_estimators': [100, 200], 'model__max_depth': [5, 10, None]},
    "GradientBoosting": {'model__n_estimators': [100, 200], 'model__learning_rate': [0.05, 0.1]}
}

best_models = {}
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

for name, model in models.items():
    print(f"\n--- Training {name} ---")
    pipe = Pipeline(steps=[('preprocessor', preprocessor), ('model', model)])
    grid = GridSearchCV(pipe, param_grids[name], cv=cv, scoring='roc_auc', n_jobs=-1)
    grid.fit(X_train, y_train)
    best_models[name] = grid.best_estimator_
    print(f"Best params: {grid.best_params_}")
    print(f"Best CV ROC-AUC: {grid.best_score_:.4f}")
```

## Evaluation

```python

for name, model in best_models.items():
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:,1]
    
    print(f"\nModel: {name}")
    print("ROC-AUC:", roc_auc_score(y_test, y_proba))
    print(classification_report(y_test, y_pred))
    
    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f"Confusion Matrix - {name}")
    plt.show()
```

## Obserations after training

After training the model had very strong preformance across the board, with random forest out preforming the other models. There could be marginal improvements on the churn prediction side, but as a startiing point 77% is not too bad! 98% accuracy overall is pretty good, and as long as the data being evaluated has similar columns as the test set of data it should make fairly reliable predictions towards customer churn. 

The main drivers of retention are the customers getting cash back from their purchases as well as coupon use, so targeting false churn risks doesn't hold too much risk. This also is due to coupons having a strong positive correlation to increasing amount of orders by customers. Manual evaluation of the targeted users of churn risk would be fairly easy to tell false churn, since the team could look at these two factors to determine if the targeted users are actually at risk.

## Best Preforming Model (Random Forest) Visualizations

```python
best_rf = best_models['RandomForest']
# Calculate ROC curve points and AUC
y_pred_proba = best_rf.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)
#Plot
plt.figure(figsize=(7,6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f"ROC curve (AUC = {roc_auc:.3f})")
plt.plot([0, 1], [0, 1], color='gray', linestyle='--', lw=1)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver Operating Characteristic (ROC) Curve")
plt.legend(loc="lower right")
plt.grid(alpha=0.3)
plt.show()

#calculate precision, recall, and average precision
precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
avg_precision = average_precision_score(y_test, y_pred_proba)
# Plot
plt.figure(figsize=(7,6))
plt.plot(recall, precision, color='blue', lw=2, label=f"Avg Precision = {avg_precision:.3f}")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.legend(loc="lower left")
plt.grid(alpha=0.3)
plt.show()

```

## Hyperparameter Tuning

```python
rf = RandomForestClassifier(random_state=42)

param_grid = {
    'model__n_estimators': [200, 400],
    'model__max_depth': [None, 10, 20],
    'model__min_samples_split': [2, 5],
    'model__min_samples_leaf': [1, 2],
    'model__max_features': ['sqrt', 'log2', None]
}

pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('model', rf)
])

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

grid_search = GridSearchCV(
    pipe, 
    param_grid=param_grid, 
    cv=cv, 
    scoring='roc_auc',
    n_jobs=-1,
    verbose=2
)
grid_search.fit(X_train, y_train)
print("Best parameters:", grid_search.best_params_)
print("Best CV ROC-AUC:", grid_search.best_score_)
```

```python
best_rf = grid_search.best_estimator_
y_pred = best_rf.predict(X_test)
y_proba = best_rf.predict_proba(X_test)[:,1]
print("Test ROC-AUC:", roc_auc_score(y_test, y_proba))
print(classification_report(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix - Random Forest")
plt.show()
```

## Final Conclusions

After hyperparameter tuning the random forest model there were very marginal improvements. The model was already preforming strongly, but the tuning yielded 1% improvement of overall accuracy. Recall for predicting churn managed to get a 2% increase as well. The model will be able to predict churn given the columns in a new set of data, minus the 'Churn' column were the same, and only need manual review for 23% of the candidates to ensure users weren't getting too much cashback and coupons to affect bottom line.
